```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Type of Plots

## Line Plots

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
```

```python
plt.plot(x, y);
plt.title("Simple line plot");
plt.xlabel("X-axis label");
plt.ylabel("Y-axis label");
plt.grid(True);
```

## Bar Chart

```python
months = ['jan', 'feb', 'mar', 'apr']
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize=(10,6))
plt.bar(months, revenue)
plt.title("Monthly revenue Bar chart");
plt.xlabel("Months");
plt.ylabel("Revenue");
```

## Subplots && Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o', label="Original price");
plt.plot(days, predicted_price, 'r-o', label="Predicted price");
plt.legend();
plt.title("Plot with legends");
plt.xlabel("weekdays");
plt.ylabel("price");
```

### Subplots

```python
# First plot - Original Price
plt.subplot(3, 1, 1);
plt.plot(days, original_price, 'b-o', label="Original price");
plt.title("Original prices over time");
plt.xlabel("weekdays");
plt.ylabel("price");
plt.legend();

# Second plot - Predicted Price
plt.subplot(3, 1, 2);
plt.plot(days, predicted_price, 'r-o', label="Predicted price");
plt.title("Predicted prices over time");
plt.xlabel("weekdays");
plt.ylabel("price");
plt.legend();

#Third plot - Difference between Original and Predicted Prices
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3);
plt.bar(days, price_difference, color='purple', label="Price Difference");
plt.title("Difference between Original price & Predicted price");
plt.xlabel("days");
plt.ylabel("price difference");
plt.legend();

plt.tight_layout()
```

## saving the figure

```python
plt.plot(days, original_price, 'b-o', label="Original price");
plt.plot(days, predicted_price, 'r-o', label="Predicted price");
plt.legend();
plt.title("Plot with legends");
plt.xlabel("weekdays");
plt.ylabel("price");

plt.savefig("example_prediction.png", dpi=300)
```
